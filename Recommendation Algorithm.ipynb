```python
#Import libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from itertools import product
```

```python
#Read dataset
df = pd.read_csv('output_1.csv') 
#df = df_original[0:127]
id = pd.DataFrame(df.index + 1)
df.insert (0, "Id", id)
df['age'] = df['age'].map(str)
df['Id'] = df['Id'].map(str)
df
```

```python
#Count number of rows and columns in the data set
df.shape
```

```python
#Create a list of important colums for the recommendation engine
columns = ['age', 'Interests', 'Looking for', 'Language']
```

```python
#Show the data
df[columns].head(3)
```

```python
#Check for missing values in important columns
df[columns].isnull().values.any()
```

```python
event = ["Sports", "Concerts", "Courses", "Restaurants", "Museums", "Dinnerparty", "Zoo", "Counseling"]
language = ["English", "Russian", "Spanish", "German"]
age_group = ["0-12", "12-18", "21-30", "30-80"]

events = list(product(event,language, age_group))
events
df_events = pd.DataFrame (events, columns = ['Event', "Language", "Age Group"])
df_events
```

```python
#Create a function to combine the values of the important columns into a single string of Output1
def get_important_features(data):
    important_features = []
    for i in range(0, df.shape[0]):
        important_features.append(data['Interests'][i]+''+data['Looking for'][i]+''+data['Language'][i]+data['age'][i])
        
    return important_features
```

```python
#Create column to hold the combined strings
df['important features'] = get_important_features(df)

# Show the data
df.head(3)
```

```python
#Convert the text to a matrix of taken counts
cm = CountVectorizer().fit_transform(df['important features'])
```

```python
#Get the cosine similarity matrix from the count matrix
cs = cosine_similarity(cm)
#Print the cosine similarity matrix
print(cs)
```

```python
cs.shape
```

```python
title = "coding"

#Find the id
user_id = df[df.Interests == title]["Id"].values[0]

```

```python
#Create a list of enumerations for the similarity score
scores = list(enumerate(cs[user_id]))
```
